```python
#Import Library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

#Baca dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/PunyaAneNihAlfaRohamn/Komputer-Grafis/main/UAS/stock.csv')
dataset.head(13)
```

# grafik histogram atribut Purchase Price, Current Price dan Gain/Loss

```python
plt.figure(figsize=(25, 10))

plt.subplot(1, 3, 1)
sns.histplot(dataset['Purchase Price ($)'], kde=True, bins=20, edgecolor='w', color="k")
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(dataset['current Price ($)'], kde=True, bins=20, edgecolor='k', color="y")
plt.xlabel('current Price')
plt.ylabel('Frequency')
plt.title('Histogram of current Price')

plt.subplot(1, 3, 3)
sns.histplot(dataset['Gain/Loss ($)'], kde=True, bins=20, edgecolor='w', color="r")
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

plt.tight_layout()
plt.show()
```

# visualisasi data untuk menampilkan Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi

```python
#Distribusi dataset pada setiap atribut Purchase Price, Current Price, dan Gain/Loss dengan grafik histogram
#Mengurutkan data berdasarkan dari Dividend / Shares
data = dataset.sort_values(by="Dividend/share", ascending=False)

#mendapatkan 5 Dividen/Shares tertinggi
top_5 = data.head(5)

#Plot the Data
plt.figure(figsize=(12, 5))
sns.barplot(x="Stock Name", y="Dividend/share", data=top_5)
plt.title("5 Saham Teratas dengan Dividen/share Tertinggi")
plt.show()
```

# visualisasi data untuk menampilkan Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi

```python
#Mengurutkan data berdasarkan dari Market Value
data = dataset.sort_values(by="Market Value ($)", ascending=False)

#Mendapatkan 5 Market Value tertinggi
top_5 = data.head(5)

# Plot the data
plt.figure(figsize=(12, 5))
sns.barplot(x="Stock Name", y="Market Value ($)", data=top_5)
plt.title("5 Saham Teratas dengan Nilai Pasar Tertinggi")
plt.show()
```

# visualisasi data untuk menampilkan Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi

```python
#Mengurutkan data berdasarkan dari Annual Yield (%)
data = dataset.sort_values(by="Annual Yield (%)", ascending=False)

#Mendapatkan 7 Annual Yield tertinggi
top_7 = data.head(7)

# Plot the data
plt.figure(figsize=(12, 5))
sns.barplot(x="Stock Name", y="Annual Yield (%)", data=top_7)
plt.title("7 Saham Teratas dengan Hasil Tahunan Tertinggi ")
plt.show()
```
